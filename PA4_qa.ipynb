# Part 1: K-Means Clustering

```python
import matplotlib.pyplot as plt
from kmeans import load_data, test_pytorch, test_sckitlearn
```

```python
train_data, test_data = load_data()
ks = [1, 2, 3, 4, 5]
for k in ks:
    plt.figure(0)
    plt.suptitle(f"PyTorch vs. SKLearn K-Means Clustering, K={k}")
    plt.subplot(121).set_title("PyTorch")
    print(f"K-Means Clustering results for K={k}")
    test_pytorch(train_data, test_data, k=k)  # plots clustering using PyTorch method
    plt.subplot(122).set_title("SKLearn")
    test_sckitlearn(train_data, test_data, k=k)  # plots clustering using SKLearn method
    plt.show()
```

### Comments
I notice that if I run the test script multiple times, the clustering results vary. Specifically, the data points that are on a border between two clusters seem to randomly oscillate between the two's ownership. I notice this random behaviour *only* occurs in the PyTorch implementation, *not* the SKLearn. This is due to the random initialization of the cluster means in comparison to the SKLearn implementation where we decide the NumPy random seed.
